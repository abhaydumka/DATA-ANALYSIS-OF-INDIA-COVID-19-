## Exploratory Analysis and Visualization / Asking and Answering Questions

In this section I have imported two libraries seaborn and matplotlib for analysis and visualization of data. Through visualization, I am able to drawn more informed decision. Also,I have executed so many queries, through which I am able to identify the pattern in covid cases of INDIA.



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (13, 9)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### Q1: When was the first confirmed case reported in India?
To answer this question I have used the .neor Not Equal to a certain value, in this case 0, and it can be applied to the new_cases series.

```python
first_confirmed_case = covid_india_df.date.loc[covid_india_df['new_cases'].ne(0).idxmax()]
```

```python
print("The first confirmed case in India was reported on {}.".format(str(first_confirmed_case)))
```

### Q2: How many cases were reported by the time the lock down was in implemented (25 March, 2020)?

```python
# Date when the locked down was implemented
locked_down_date = '2020-03-25'
```

```python
total_cases_locked_down_date = covid_india_df.total_cases.loc[covid_india_df.loc[covid_india_df.date == locked_down_date].index.values.astype(int)[0]]
```

```python
print("The total number of confirmed cases by {} was {}.".format(locked_down_date, int(total_cases_locked_down_date)))
```

### Q3: How many deaths were reported by the time the locked down was implemented in India?

```python
total_deaths_locked_down_date = covid_india_df.total_deaths.loc[covid_india_df.loc[covid_india_df.date == locked_down_date].index.values.astype(int)[0]]
```

```python
print("The total number of deaths by {} was {}.".format(locked_down_date, int(total_deaths_locked_down_date)))
```

### Q4: When was the first death related to COVID-19 reported?

```python
first_death = covid_india_df.date.loc[covid_india_df['new_deaths'].ne(0).idxmax()]
```

```python
print("The first death related to COVID-19 in India was reported on {}.".format(str(first_death)))
```

### Q5: How did the number of daily new cases and new deaths evolved in India?
To answer this question I have plotted the information.

```python
plt.plot(covid_india_df.new_cases, 'b-', linewidth=1, markersize=5, markeredgewidth=2, markeredgecolor='navy')
plt.plot(covid_india_df.new_deaths, 'r--', linewidth=1, markersize=5,)
plt.title('Daily Positve Cases vs Daily Deaths')
plt.legend(['Daily Cases', 'Daily Deaths'])
plt.xlabel('Days'); plt.ylabel('Count');
```

Through this graph we can say that at the early stages of the locked down, the new deaths reported were low, as compared to the number of cases. But after some days later a big growth in cases recored.

### Q6: Which days were the ones that reported the highest number of cases in India?
To answer this i have considered the average of daily cases.

```python
covid_india_df.new_cases.describe()
```

I have considered 1500 cases as threshold, and created a new data frame to analyze the data of these days.

```python
high_positive_cases_df = covid_india_df[covid_india_df.new_cases > 1500]
high_positive_cases_df
```

```python
high_positive_cases_df.shape
```

Now, I have sorted these information based on the highest confirmed cases date.

```python
high_positive_cases_df.sort_values('new_cases', ascending=False)
```

And now I have ploted this information:

```python
plt.figure(figsize=(30,30))
sns.barplot(high_positive_cases_df.new_cases, high_positive_cases_df.date)
plt.title('Days with the Highest Positive Cases')
plt.xlabel('Number of Daily Positive Cases'); plt.ylabel('Date');
```

The day with the most reported cases was May 7th, with over 414188 cases.

### Q7: Which days were the ones that reported over 500 deaths?

```python
high_deaths_days_df = covid_india_df[covid_india_df.new_deaths > 500]
high_deaths_days_df
```

```python
sns.barplot(high_deaths_days_df.new_deaths, high_deaths_days_df.date)
plt.title('Days with Over 500 Deaths')
plt.xlabel('Deaths'); plt.ylabel('Date');
```

This graphs tells that 10 June was the day with the highest deaths reported, with 6148 deaths.

### Q8: How many positive cases did India had per month?
To answer this question, I have grouped the information by months instead of per days. To do so, I have used the .groupby() method from the Pandas Library.

```python
covid_india_month_df = covid_india_df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum()
covid_india_month_df
```

```python
sns.barplot(covid_india_month_df.index.drop(12), covid_india_month_df.new_cases.drop(index=12))
plt.title('Number of Positive Cases per Month in India')
plt.legend(['1-Jan', '2-Feb', '3-Mar', '4-Apr', '5-May', '6-June', '7-Jul', '8-Aug', '9-Sep', '10-Oct', '11-Nov', '12-Dec'],)
plt.xlabel('Month'); plt.ylabel('Number of Cases');
```

Through this graph I have observed that May was the month with the most positive cases so far.

### Q9: Which month is the one with the most number of deaths?

```python
sns.barplot(covid_india_month_df.index.drop(12), covid_india_month_df.new_deaths.drop(index=12))
plt.title('Number of Deaths Per Month in India')
plt.legend(['1-Jan', '2-Feb', '3-Mar', '4-Apr', '5-May', '6-June', '7-Jul', '8-Aug', '9-Sep', '10-Oct', '11-Nov', '12-Dec'],)
plt.xlabel('Month'); plt.ylabel('Deaths');
```

May was the month with the most deaths in India.

### Q10: How many tests have been performed by month?

```python
sns.barplot(covid_india_month_df.index.drop(12), covid_india_month_df.new_tests.drop(index=12))
plt.title('Number of Tests per Month in India')
plt.legend(['1-Jan', '2-Feb', '3-Mar', '4-Apr', '5-May', '6-June', '7-Jul', '8-Aug', '9-Sep', '10-Oct', '11-Nov', '12-Dec'],)
plt.xlabel('Month'); plt.ylabel('Tests');
```

Here, In month of August the highest number of tests were performed so far. 

### Q11: What is the positive case and death rate per month?
For this, I have created 2 new columns in the monthly data frame and named as death_rate and test_positive_rate.

```python
covid_india_month_df['death_rate'] = covid_india_month_df.new_deaths / covid_india_month_df.new_tests
covid_india_month_df['test_positive_rate'] = covid_india_month_df.new_cases / covid_india_month_df.new_tests
covid_india_month_df
```

```python
plt.plot(covid_india_month_df.index, covid_india_month_df.death_rate, 'b-x', linewidth=4, markersize=12, markeredgewidth=4, markeredgecolor='navy')
plt.plot(covid_india_month_df.index, covid_india_month_df.test_positive_rate, 'r--o', linewidth=4, markersize=12,)
plt.title('Death and Test Postive Rates per Month')
plt.legend(['Death Rate', 'Test Positive'])
plt.xlabel('Month'); plt.ylabel('Rate');
```

### Q12: What is tha porcentage of the population that have tested positive in the country?
To answer this question I have included the population of the country in the covid_india_df data frame. Also, I have created a pie chart to saw the whole scenario of covid_19 pandemic in India.

```python
covid_india_df['population'] = India_df.population
covid_india_df
```

Now, I have extracted the following information:

```python
population = covid_india_df.population[0]
total_cases = covid_india_df.total_cases[covid_india_df.total_cases.last_valid_index()]
total_deaths = covid_india_df.total_deaths[covid_india_df.total_deaths.last_valid_index()]
total_tests = covid_india_df.total_tests[covid_india_df.total_tests.last_valid_index()]
total_non_tested = population - total_tests
```

And finally, I have ploted the inforamtion.

```python
plt.pie([total_non_tested, total_cases,  total_tests, total_deaths], 
        labels=['Population Not Tested', 'Positive Cases', 'Population Tested', 'Deaths'], 
        autopct='%1.3f%%', startangle=180, counterclock=True, data=True, )
plt.title('Population in India per Tests, Cases, Deaths')
```

Through, this pie chart, I got the following results:
38.785% of the population has been tested.
2.351% of the population has been tested positive.
0.031% of the population has died.
58.833% of the population has not been tested yet.
